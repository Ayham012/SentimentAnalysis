```python
import pandas as pd
import os

raw_file_path = "C:\\Users\\ALTamimi\\Desktop\\Ayham\\Qafza\\SentimentAnalysisProject\\DataSet\\raw\\IMDB Dataset.csv"  # Update with your actual filename
processed_file_path = "C:\\Users\\ALTamimi\\Desktop\\Ayham\\Qafza\\SentimentAnalysisProject\\DataSet\\processed\\imdb_cleaned.csv"

os.makedirs("processed", exist_ok=True)

df = pd.read_csv(raw_file_path)

print("Initial Data Shape:", df.shape)
print(df.info())

df = df.drop_duplicates()

df = df.dropna(thresh=df.shape[1] * 0.6)  # Keeping rows with at least 60% valid data

for col in df.select_dtypes(include=['number']).columns:
    df[col].fillna(df[col].median(), inplace=True)

df.columns = df.columns.str.lower().str.replace(" ", "_")

if 'release_year' in df.columns:
    df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').fillna(0).astype(int)

df.to_csv(processed_file_path, index=False)

print("Processing complete! Cleaned data saved to:", processed_file_path)
print("Final Data Shape:", df.shape)
```
